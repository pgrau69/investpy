```python
from google.colab import drive
import pandas as pd

# Montar Google Drive
drive.mount('/content/drive')

# Rutas a los archivos CSV en Google Drive
file_path_ask = '/content/drive/MyDrive/Colab Notebooks/DATA/historical_ask_data_3years_3mins.csv'
file_path_bid = '/content/drive/MyDrive/Colab Notebooks/DATA/historical_bid_data_3years_3mins.csv'

# Leer los archivos CSV en DataFrames de Pandas
data_ask = pd.read_csv(file_path_ask, delimiter=',')
data_bid = pd.read_csv(file_path_bid, delimiter=',')

# Asegurarse de que todos los valores en la columna 'date' sean cadenas
data_ask['date'] = data_ask['date'].astype(str)
data_bid['date'] = data_bid['date'].astype(str)

# Eliminar la información de la zona horaria de la columna 'date'
data_ask['date'] = data_ask['date'].str.replace(r'-\d{2}:\d{2}$', '', regex=True)
data_bid['date'] = data_bid['date'].str.replace(r'-\d{2}:\d{2}$', '', regex=True)

# Convertir la columna 'date' a tipo datetime
data_ask['date'] = pd.to_datetime(data_ask['date'], errors='coerce')
data_bid['date'] = pd.to_datetime(data_bid['date'], errors='coerce')

# Configurar la columna 'date' como índice
data_ask.set_index('date', inplace=True)
data_bid.set_index('date', inplace=True)

# Renombrar la columna 'date' a 'Datetime'
data_ask.index.name = 'Datetime'
data_bid.index.name = 'Datetime'

# Eliminar las columnas 'volume', 'average' y 'barCount'
data_ask.drop(columns=['volume', 'average', 'barCount'], inplace=True)
data_bid.drop(columns=['volume', 'average', 'barCount'], inplace=True)

# Eliminar filas con fechas inválidas
data_ask.dropna(inplace=True)
data_bid.dropna(inplace=True)

# Renombrar las columnas para diferenciarlas
data_ask.columns = ['Ask_Open', 'Ask_High', 'Ask_Low', 'Ask_Close']
data_bid.columns = ['Bid_Open', 'Bid_High', 'Bid_Low', 'Bid_Close']

# Combinar los DataFrames de BID y ASK
data_combined = pd.concat([data_ask, data_bid], axis=1)

# Convertir las fechas de filtro a tipo datetime
start_date_close = pd.to_datetime('2024-05-25')
end_date_close = pd.to_datetime('2024-06-04 23:59:59')

# Filtrar el DataFrame por fechas utilizando la columna de índice 'Datetime'
data_eur = data_combined[(data_combined.index >= start_date_close) & (data_combined.index <= end_date_close)]

# Filtrar por el rango de tiempo deseado
data_eur = data_eur.between_time('02:00', '12:00')

data_eur

```

```python
import plotly.graph_objects as go

# Función para calcular velas Heiken Ashi
def heiken_ashi(data):
    ha_data = data.copy()

    ha_open = [(data['Ask_Open'][0] + data['Ask_Close'][0]) / 2]
    ha_close = [(data['Ask_Open'][0] + data['Ask_High'][0] + data['Ask_Low'][0] + data['Ask_Close'][0]) / 4]

    for i in range(1, len(data)):
        ha_close.append((data['Ask_Open'][i] + data['Ask_High'][i] + data['Ask_Low'][i] + data['Ask_Close'][i]) / 4)
        ha_open.append((ha_open[i-1] + ha_close[i-1]) / 2)

    ha_data['HA_Open'] = ha_open
    ha_data['HA_Close'] = ha_close
    ha_data['HA_High'] = ha_data[['HA_Open', 'HA_Close', 'Ask_High']].max(axis=1)
    ha_data['HA_Low'] = ha_data[['HA_Open', 'HA_Close', 'Ask_Low']].min(axis=1)

    return ha_data

# Leer los datos combinados de BID y ASK

# Calcular Heiken Ashi para el DataFrame 'data_eur'
data_eur = heiken_ashi(data_eur)

# Crear una columna para las señales de compra
data_eur['entries'] = 0

# Condición para la señal de compra: cambio de vela bajista a alcista
for i in range(1, len(data_eur)-1):
    if data_eur['HA_Close'][i] > data_eur['HA_Open'][i] and data_eur['HA_Close'][i-1] < data_eur['HA_Open'][i-1]:
        data_eur.at[data_eur.index[i+1], 'entries'] = 1

# Crear un diccionario de entradas a partir del DataFrame
entries = data_eur['entries'].to_dict()

# Crear la figura de Plotly para visualizar las señales
fig = go.Figure()

# Añadir velas Heiken Ashi
fig.add_trace(go.Candlestick(
    x=data_eur.index,
    open=data_eur['HA_Open'],
    high=data_eur['HA_High'],
    low=data_eur['HA_Low'],
    close=data_eur['HA_Close'],
    name='Heiken Ashi'
))

# Añadir señales de compra
buy_signals = data_eur[data_eur['entries'] == 1]
fig.add_trace(go.Scatter(
    x=buy_signals.index,
    y=buy_signals['HA_Low'],
    mode='markers',
    marker=dict(symbol='triangle-up', color='green', size=10),
    name='Buy Signal'
))

# Configurar el layout del gráfico
fig.update_layout(
    title='Heiken Ashi con Señales de Compra',
    xaxis_title='Fecha',
    yaxis_title='Precio',
    xaxis_rangeslider_visible=False
)

# Mostrar el gráfico
fig.show()
```

```python
import pandas as pd
import numpy as np
import time

# Función para manejar una entrada larga
def handle_long_entry(capital, risk_percentage_long, entry_price, bid_low, bid_high):
    stop_loss_long = bid_low  # Usar el Bid_Low para Stop Loss
    take_profit_long = bid_high  # Usar el Bid_High para Take Profit
    position_size_long = capital * risk_percentage_long / (entry_price - stop_loss_long)
    return stop_loss_long, take_profit_long, position_size_long

# Función para manejar una orden de venta stop (short)
def handle_sell_stop_order(bid_low, ask_high, ask_low):
    sell_stop_price = bid_low  # Usar el Bid_Low como precio de entrada para la orden de venta
    stop_loss_short = ask_high  # Usar el Ask_High para Stop Loss
    take_profit_short = ask_low  # Usar el Ask_Low para Take Profit
    return sell_stop_price, stop_loss_short, take_profit_short

# Función para procesar el cierre de una posición larga
def process_long_exit(low_price, high_price, stop_loss_long, take_profit_long, capital, risk_percentage_long, timestamp, operation_id, entry_price):
    capital_change = 0
    if low_price <= stop_loss_long:
        capital_change = -capital * risk_percentage_long
        capital += capital_change
        long_stop_loss_timestamps.append((timestamp, operation_id, stop_loss_long, take_profit_long, entry_price))
        print(f"{timestamp} Long - SALIDA por Stop Loss en operación {operation_id}: Stop Loss alcanzado: {stop_loss_long} Cambio de capital al salir: {capital_change}")
    elif high_price >= take_profit_long:
        capital_change = capital * risk_percentage_long
        capital += capital_change
        long_take_profit_timestamps.append((timestamp, operation_id, stop_loss_long, take_profit_long, entry_price))
        print(f"{timestamp} Long - SALIDA por Take Profit en operación {operation_id}: Take Profit alcanzado: {take_profit_long} Cambio de capital al salir: +{capital_change}")
    return capital, capital_change

# Función para procesar el cierre de una posición corta
def process_short_exit(high_price, low_price, stop_loss_short, take_profit_short, capital, risk_percentage_short, timestamp, operation_id, entry_price):
    capital_change = 0
    if high_price >= stop_loss_short:
        capital_change = -capital * risk_percentage_short
        capital += capital_change
        short_stop_loss_timestamps.append((timestamp, operation_id, stop_loss_short, take_profit_short, entry_price))
        print(f"{timestamp} Short - SALIDA por Stop Loss en operación {operation_id}: Stop Loss alcanzado: {stop_loss_short} Cambio de capital al salir: {capital_change}")
    elif low_price <= take_profit_short:
        capital_change = capital * risk_percentage_short
        capital += capital_change
        short_take_profit_timestamps.append((timestamp, operation_id, stop_loss_short, take_profit_short, entry_price))
        print(f"{timestamp} Short - SALIDA por Take Profit en operación {operation_id}: Take Profit alcanzado: {take_profit_short} Cambio de capital al salir: +{capital_change}")
    return capital, capital_change

# Función principal para ejecutar la estrategia
def run_strategy(data_eur, entries, capital, risk_percentage_long, risk_percentage_short):
    position_open = False
    short_position_open = False
    capital_over_time = []  # Lista para almacenar el capital a lo largo del tiempo
    capital_changes = []
    global long_position_timestamps
    global short_position_timestamps
    global long_take_profit_timestamps
    global short_take_profit_timestamps
    global long_stop_loss_timestamps
    global short_stop_loss_timestamps
    global sell_stop_price
    long_position_timestamps = []
    short_position_timestamps = []
    long_take_profit_timestamps = []
    short_take_profit_timestamps = []
    long_stop_loss_timestamps = []
    short_stop_loss_timestamps = []
    sell_stop_triggered_timestamps = []
    operation_id = 0  # Inicializar identificador de operación

    for timestamp, signal in entries.items():
        if signal:
            if not position_open:
                operation_id += 1  # Incrementar identificador de operación
                entry_price = data_eur['Ask_Open'].loc[timestamp]
                bid_low = data_eur['Bid_Low'].loc[timestamp]
                bid_high = data_eur['Bid_High'].loc[timestamp]
                stop_loss_long, take_profit_long, position_size_long = handle_long_entry(
                    capital, risk_percentage_long, entry_price, bid_low, bid_high
                )
                position_open = True
                long_position_timestamps.append((timestamp, operation_id, stop_loss_long, take_profit_long, entry_price))
                print(f"{timestamp} Long - Valor de entrada en operación {operation_id}: {entry_price} Stop Loss asignado: {stop_loss_long} Take Profit asignado: {take_profit_long} Capital al entrar: {capital}")

                ask_low = data_eur['Ask_Low'].loc[timestamp]
                ask_high = data_eur['Ask_High'].loc[timestamp]
                sell_stop_price, stop_loss_short, take_profit_short = handle_sell_stop_order(
                    bid_low, ask_high, ask_low
                )
                short_position_open = False
                print(f"{timestamp} Sell Stop Order colocada en operación {operation_id} a: {sell_stop_price} con SL: {stop_loss_short} y TP: {take_profit_short}")

        if position_open or short_position_open:
            low_price = data_eur['Bid_Low'].loc[timestamp]
            high_price = data_eur['Bid_High'].loc[timestamp]

            if position_open and not short_position_open:
                if low_price <= sell_stop_price:
                    short_position_open = True
                    sell_stop_triggered_timestamps.append((timestamp, operation_id, stop_loss_short, take_profit_short, sell_stop_price))
                    print(f"{timestamp} Sell Stop Order activada en operación {operation_id} al precio: {sell_stop_price}")

            if position_open:
                capital, capital_change = process_long_exit(
                    low_price, high_price, stop_loss_long, take_profit_long, capital, risk_percentage_long, timestamp, operation_id, entry_price
                )
                if capital_change != 0:
                    position_open = False
                    capital_changes.append(capital_change)

            if short_position_open:
                ask_high = data_eur['Ask_High'].loc[timestamp]
                ask_low = data_eur['Ask_Low'].loc[timestamp]
                capital, capital_change = process_short_exit(
                    ask_high, ask_low, stop_loss_short, take_profit_short, capital, risk_percentage_short, timestamp, operation_id, sell_stop_price
                )
                if capital_change != 0:
                    short_position_open = False
                    capital_changes.append(capital_change)

            capital_over_time.append(capital)

    # Convertir la lista de capital a un DataFrame para cálculos de drawdown
    capital_df = pd.DataFrame(capital_over_time, columns=['Capital'])
    capital_df['Peak'] = capital_df['Capital'].cummax()
    capital_df['Drawdown'] = capital_df['Capital'] - capital_df['Peak']
    capital_df['Drawdown_Percentage'] = (capital_df['Drawdown'] / capital_df['Peak']) * 100

    # Calcular el máximo drawdown y su duración
    max_drawdown = capital_df['Drawdown_Percentage'].min()
    max_drawdown_duration = (capital_df['Drawdown_Percentage'] < 0).astype(int).groupby(capital_df['Drawdown_Percentage'].ge(0).astype(int).cumsum()).cumsum().max()

    print(f"Max Drawdown: {max_drawdown}%")
    print(f"Max Drawdown Duration: {max_drawdown_duration} periods")

    return capital_over_time, capital_changes, long_position_timestamps, short_position_timestamps, long_take_profit_timestamps, short_take_profit_timestamps, long_stop_loss_timestamps, short_stop_loss_timestamps, sell_stop_triggered_timestamps

# Parámetros de la estrategia
capital = 100.0
risk_percentage_long = 0.01
risk_percentage_short = 0.02

# Medir el tiempo de ejecución
start_time = time.time()

# Ejecutar la estrategia
capital_over_time, capital_changes, long_position_timestamps, short_position_timestamps, long_take_profit_timestamps, short_take_profit_timestamps, long_stop_loss_timestamps, short_stop_loss_timestamps, sell_stop_triggered_timestamps = run_strategy(
    data_eur, entries, capital, risk_percentage_long, risk_percentage_short
)

end_time = time.time()
execution_time = end_time - start_time

print(f"Tiempo de ejecución: {execution_time:.4f} segundos")
```
